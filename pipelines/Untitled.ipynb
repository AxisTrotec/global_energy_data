```python
df = pd.read_csv(prefix, sep=',', encoding='utf-8', dtype=dtype, nrows=100)
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql+psycopg://root:root@localhost:5432/energy_data')
```

```python
print(pd.io.sql.get_schema(df, name='energy_data', con=engine))
```

```python
df.head(n=0).to_sql(name='energy_data', con=engine, if_exists='replace')
```

```python
df_iter = pd.read_csv(
    prefix,
    dtype=dtype,
    iterator=True,
    chunksize=100000
)
```

```python
from tqdm.auto import tqdm

first = True

for df_chunk in tqdm(df_iter):

    if first:
        # Create table schema (no data)
        df_chunk.head(0).to_sql(
            name="energy_data",
            con=engine,
            if_exists="replace"
        )
        first = False
        print("Table created")

    # Insert chunk
    df_chunk.to_sql(
        name="energy_data",
        con=engine,
        if_exists="append"
    )

    print("Inserted:", len(df_chunk))
```
